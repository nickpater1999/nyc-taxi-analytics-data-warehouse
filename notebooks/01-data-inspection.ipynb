```python
# Objective:
    # Inspect the raw data for this project, consisting of 12 parquet files
    # Record discrepancies between column names and data types across files
```

```python
# Import libraries
import os
import glob
import pandas as pd
```

```python
# Define where to find the raw data, found in the form of 12 .parquet files
```

```python
# Relative to the location of this notebook file, raw data is found in 'raw_data/*.parquet'

# Generate and sort 12 unique strings representing the relatively located raw data file names
data_dir = "raw_data"
parquet_pattern = os.path.join(data_dir, "*.parquet")
parquet_files = sorted(glob.glob(parquet_pattern))
#print(parquet_files)
```

```python
# Inspect .parquet files: what are column names and their datatypes?
```

```python
# Initialize dictionary: used to store column data types per file
col_types_per_file = {}

# Iterate over all parquet files and collect column types
for file in parquet_files:
    df = pd.read_parquet(file)
    for col in df.columns:
        if col not in col_types_per_file:
            col_types_per_file[col] = {}  # dictionary mapping file -> dtype
        col_types_per_file[col][os.path.basename(file)] = str(df[col].dtype)

# Print results in a readable way
print("Columns and their data types across all files:")
for col, file_dict in col_types_per_file.items():
    print(f"\nColumn: {col}")
    for fname, dtype in file_dict.items():
        print(f"  {fname}: {dtype}")
```

```python
# Conclusions from this inspection:
    # 2023-01 parquet file has different column VendorID datatype
    # 2023-01 parquet file has different column PULocationID datatype
    # 2023-01 parquet file has different column DOLocationID datatype
    # 2023-01 parquet file has airport_fee column, not Airport_fee column
```
